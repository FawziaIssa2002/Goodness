```python
import re
from pyarabic.araby import tokenize, strip_tashkeel, strip_tatweel, normalize_ligature
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def clean_arabic_text(text):
    text = str(text)
    
    # Remove non-Arabic characters
    text = re.sub("[^\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF\uFB50-\uFDCF\uFDF0-\uFDFF\uFE70-\uFEFF]+", " ", text)
    
    # Remove tatweel (elongation) characters ุงูุชูุฑุงุฑ
    text = strip_tatweel(text)
    
    # Remove diacritical marks (tashkeel) characters ุงูุชุดููู
    text = strip_tashkeel(text)
    
     # Remove English characters and numbers
    text = re.sub('[A-Za-z0-9]',' ',text)
    
    # Remove multiple spaces and newline characters
    text = re.sub(r'\s+', ' ', text)
    text = re.sub('\n', ' ', text)
    
    # Normalize alef maksura (ู) to ุง
    text = re.sub(r'ู', 'ุง', text)
    
    # Normalize teh marbuta (ุฉ) to ู
    text = re.sub(r'ุฉ', 'ู', text)
    
    # Normalize hamza (ุก) and alif hamza (ุฃุฅุข) to ุง
    text = re.sub("[ุฅุฃุขูุก]", "ุง", text)
    
    # Normalize waw hamza (ุค) to ู
    text = re.sub(r'ุค', 'ู', text)
    
    # Remove duplicate letters
    text = re.sub(r'(.)\1+', r'\1', text)
    
    # Remove punctuation
    text = re.sub(r'[^\w\s]', ' ', text)
       
    # Normalize spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenize the text into words
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('arabic'))

    filtered_words = [word for word in tokens if word.casefold() not in stop_words]
    
    # Replace emojis
    emojis = {
        ":)": "smile",
        ":(": "sad"
        # Add more emoji replacements as needed
    }  
    filtered_words = [emojis.get(token, token) for token in filtered_words]
    
    return filtered_words
```
